```python
import csv, sqlite3
import os
import sqlite3
from sqlite3 import Error
import gzip


def create_connection(db_file, delete_db=False):
    if delete_db and os.path.exists(db_file):
        os.remove(db_file)
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        conn.execute("PRAGMA foreign_keys = 1")
    except Error as e:
        print(e)

    return conn

db_file = 'project.db'
conn = create_connection(db_file, delete_db=True)



```

```python
def piegraph(df2, title):
    labels = 'VioN', 'MajorN', 'OthN', 'PropN', 'NoCrimN'
    sizes = [int(df2['VioN']), int(df2['MajorN']), int(df2['OthN']), int(df2['PropN']), int(df2['NoCrimN'])]

    fig1, ax1 = plt.subplots()
    ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
            shadow=True, startangle=90)
    ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.title(title)
    plt.legend()

    plt.show()
```

## User applications

```python
def validateaddress(address):
    df = pd.read_sql_query(f"select Location_name from Safety;", conn)
    
    if address not in df.Location_name.values:
        cann = set()
        address = set(map(lambda addy:addy.upper(),address.split()))
        for loc in df.Location_name.values:
            locSet = set(map(lambda addy:addy.upper(),loc.split()))
            if locSet & address:
                cann.add(loc)
        if not cann:
            print(df)
            print('No similar matches')
        else:
            print('Similar Matches')
            for c in cann:
                print(c)
        return False
    return True
            
```

```python
def userinputcrime():
    
    address = input("Put in Address ")
    while validateaddress(address) == False:
        address = input("Try Again ")
    
    df = pd.read_sql_query(f"select * from Safety where location_name is '{address}';", conn)
    plotcrime(df)
    
```

```python
def userinputpie():
    address = input('Put in Address ')
    while validateaddress(address) == False:
        address = input("Try Again ")
    year = input('Put in Year')
    while validateyear(year, address) == False:
        year = input("Try Again ")
    df = pd.read_sql_query(f"select * from Safety where location_name is '{address}' and school_year is '{year}';", conn)
    piegraph(df,f'{year} {address}')
```
